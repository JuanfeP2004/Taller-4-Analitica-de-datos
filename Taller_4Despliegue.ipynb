```python
from types import SimpleNamespace
import streamlit as st

st.title('Predicción de desorden de sueño')

age = st.slider('Edad', min_value=18, max_value=80, value=30, step=1)
gender = st.selectbox('Genero', ['Male', 'Female'])
occupation = st.selectbox('Ocupacion', ['Software Engineer', 'Doctor','Nurse','Engineer','Lawyer',
                                       'Teacher','Accountant','Salesperson','Scientist'])
sleep_duration = st.slider('Duracion de sueño', min_value=0.0, max_value=24.0, value=8.0, step=0.1)
physical_activity_level = st.slider('Actividad fisica', min_value=0, max_value=300, value=30, step=1)
bmi_category = st.selectbox('Categoria de BMI', ['Normal', 'Overweight', 'Obesity'])
heart_rate = st.slider('Frecuencia cardiaca', min_value=30, max_value=200, value=60, step=1)
dbp = st.slider('Presion arterial diastolica', min_value=60, max_value=120, value=80, step=1)
daily_steps = st.slider('Pasos diarios', min_value=0, max_value=10000, value=1000, step=1)

datos = [[age, gender, occupation, sleep_duration, physical_activity_level, bmi_category, heart_rate, dbp, daily_steps]]
data = pd.DataFrame(datos, columns=['Edad', 'Genero','Ocupacion','Horas de sueño','Actividad Fisica', 'IMC', 'Frecuencia cardiaca',
                                    'Presion diastolica', 'Pasos diarios']) #Dataframe con los mismos nombres de variables
```
